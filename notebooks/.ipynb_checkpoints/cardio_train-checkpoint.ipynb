## preprocessing

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Convert age  Days to Years

```python
df['age_years']=(df['age']/365).astype(int)
```

```python
df
```

```python
df['gender']=df['gender'].map({1:'Female',2:'Male'})
```

```python
df
```

```python
df.dtypes
```

## convert weight datatype from float to int

```python
df['weight']=df['weight'].astype(int)
```

```python
df.dtypes
```

```python
(df['height']>220).sum()
```

```python
(df['height']<120).sum()
```

```python
df = df[(df["height"] >= 120) & (df["height"] <= 220)]
```

```python
df.shape
```

```python
df = df[(df["weight"] >= 30) & (df["weight"] <= 200)]
```

```python
df.shape
```

```python
df = df[(df["ap_hi"] >= 80) & (df["ap_hi"] <= 250)]
```

```python
df = df[(df["ap_lo"] >= 40) & (df["ap_lo"] <= 150)]
```
